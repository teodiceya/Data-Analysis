### –ó–∞–¥–∞—á–∞ ‚Ññ2

–î–∞–ª–µ–µ –ø–æ–π–¥—É—Ç –∑–∞–¥–∞—á–∏ –ø–æ—Å–ª–æ–∂–Ω–µ–µ; –≤—ã–≤–µ–¥–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É –¥–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏.

```python
# –≤—Ä–µ–º–µ–Ω–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏
# –∑–∞—Ç–µ–º –æ–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞–Ω–Ω—ã–µ —Å –∫–Ω–∏–∂–Ω–æ–π —Ç–∞–±–ª–∏—Ü–µ–π –ø–æ book_id

query = """
        WITH rew AS (SELECT book_id,
                            COUNT(review_id) as count_reviews
                     FROM reviews
                     GROUP BY book_id),
             rat AS (SELECT book_id,
                            AVG(rating) as average_rating
                     FROM ratings
                     GROUP BY book_id)

        SELECT b.title,
               rew.count_reviews,
               rat.average_rating
        FROM books as b
        LEFT OUTER JOIN rew ON b.book_id=rew.book_id
        LEFT OUTER JOIN rat ON b.book_id=rat.book_id
        ORDER BY average_rating DESC;
        """

pd.io.sql.read_sql(query, con = engine) 
```

### –ó–∞–¥–∞—á–∞ ‚Ññ3

–°–ª–µ–¥—É—é—â–∏–º –∑–∞–ø—Ä–æ—Å–æ–º –ø–æ—Å–º–æ—Ç—Ä–∏–º –∏–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–∞, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã–ø—É—Å—Ç–∏–ª–∏ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –∫–Ω–∏–≥, –ø–æ —Å–∞–º–∏–º –∫–Ω–∏–≥–∞–º–∏ –±—É–¥–µ—Ç –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –æ—Ç 50 —Å—Ç—Ä–∞–Ω–∏—Ü –∏ –±–æ–ª—å—à–µ.

```python
# –≤—Ä–µ–º–µ–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Å –∫–Ω–∏–≥–∞–º–∏ —Ç–æ–ª—â–∏–Ω–æ–π 50 —Å—Ç—Ä–∞–Ω–∏—Ü –∏ –±–æ–ª—å—à–µ,
# –æ–±—ä–µ–¥–∏–Ω—è–µ–º —Å –¥–∞–Ω–Ω—ã–º–∏ –æ–± –∏–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–∞—Ö –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ —É–±—ã–≤–∞–Ω–∏—é

query = """
        WITH b AS (SELECT publisher_id,
                          title
                   FROM books
                   WHERE num_pages >= 50)

        SELECT p.publisher,
               COUNT(b.title) as amount_books
        FROM publishers as p
        INNER JOIN b ON p.publisher_id=b.publisher_id
        GROUP BY p.publisher
        ORDER BY amount_books DESC;
        """

pd.io.sql.read_sql(query, con = engine) 
```

### –ó–∞–¥–∞—á–∞ ‚Ññ4

–û–ø—Ä–µ–¥–µ–ª–∏–º –∞–≤—Ç–æ—Ä–∞ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥, –ø—Ä–∏—á–µ–º –∫–Ω–∏–≥–∏ –¥–æ–ª–∂–Ω—ã –∏–º–µ—Ç—å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–æ–∫.

```python
# –≤—Ä–µ–º–µ–Ω–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–≤ –∫–Ω–∏–≥ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏
# –∑–∞—Ç–µ–º –æ–±—ä–µ–¥–∏–Ω—è–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Å –∞–≤—Ç–æ—Ä–∞–º–∏ —á–µ—Ä–µ–∑ —Ç–∞–±–ª–∏—Ü—É books –∏ —Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ

query = """
        WITH rat AS (SELECT book_id,
                            AVG(rating) as average_rating,
                            COUNT(rating) as amount_ratings
                     FROM ratings
                     GROUP BY book_id
                     HAVING COUNT(rating) >= 50)

        SELECT a.author,
               AVG(rat.average_rating) as average_rating
        FROM rat
        LEFT OUTER JOIN books AS b ON b.book_id=rat.book_id
        LEFT OUTER JOIN authors AS a ON a.author_id=b.author_id
        GROUP BY a.author
        ORDER BY average_rating DESC;
        """

pd.io.sql.read_sql(query, con = engine) 
```
